### Plot Labels and Title

```python
yearExperience = [1, 3, 5, 10, 15]
salary = [700, 900, 1300, 1800, 2500]
```

```python
plt.style.use('bmh')
plt.figure(figsize=(20, 6))
plt.plot(yearExperience, salary, marker='H', markerfacecolor='black', markersize=10, linewidth=4.5)
plt.title('Salary by Expereience', pad=40, fontdict=dict(color='Orange', fontsize=35, fontweight='semibold'))
plt.xlabel('Years Experience', labelpad=30, fontdict=dict(color='#1F8A70', fontsize=20, fontweight='heavy'))
plt.ylabel('Average Salary', rotation='horizontal', labelpad=95, fontdict=dict(color='#1F8A70', fontsize=20, fontweight='heavy'))
```

### X Y Ticks

### Bar Charts

```python
seed = ['apple', 'banana', 'orange', 'jackfruits', 'coconut']
total_dead = [30, 14, 18, 25, 10]
total_growth = [100, 120, 118, 58, 36]
```

```python
plt.bar(seed, total_dead, width=0.5, color="#F55050")
```

```python
plt.bar(seed, total_growth, width=0.5, color="#54B435")
```

```python
plt.bar(seed, total_dead, width=0.4, color="#F55050", label='Growth')
plt.bar(seed, total_growth, width=0.2, color="#54B435", align='edge', label='Died')
plt.title("Total Growth and Dead Seed", pad=20, fontdict=dict(fontweight='bold'))
plt.ylabel("Total", color="#1746A2")
plt.xlabel("Fruits", labelpad=10, color="#1746A2")
plt.legend(facecolor='#FAF7F0', frameon=True, shadow=True)
```

```python
plt.bar(seed, total_dead, width=0.4, bottom=total_growth, color="#F55050", label='Growth')
plt.bar(seed, total_growth, width=0.4, color="#54B435", label='Died')
plt.title("Total Growth and Dead Seed", pad=20, fontdict=dict(fontweight='bold'))
plt.ylabel("Total", color="#1746A2")
plt.xlabel("Fruits", labelpad=10, color="#1746A2")
plt.legend(facecolor='#FAF7F0', frameon=True, shadow=True)
```

### Pie Chart

```python
product = ['book', 'pencil', 'eraser', 'ruler', 'paper']
sell = [5.78, 12.45, 8.19, 4.98, 6.90]
```

```python
plt.pie(sell, labels=product, autopct='%1.2f%%')
plt.title("Normal Pie Chart")
```

```python
plt.pie(sell, labels=product, autopct='%1.2f%%', explode=[0, 0.15, 0, 0, 0], shadow=True)
plt.title("Explode Pie Chart")
```

### Subplots

```python
nums = np.arange(1, 6)
```

```python
plt.figure(figsize=[16, 4])
plt.suptitle("My Subplots", size=20)

plt.subplot(1, 4, 1)
plt.plot(nums, color="#FF731D", label='x', marker='p')
plt.legend()

plt.subplot(1, 4, 2)
plt.plot(nums**2, color="#1746A2", label='x^2', marker='p')
plt.legend()

plt.subplot(1, 4, 3)
plt.plot(nums**3, color="#A66CFF", label='x^3', marker='p')
plt.legend()

plt.subplot(1, 4, 4)
plt.plot(nums**4, color="#31E1F7", label='x^4', marker='p')
plt.legend()

plt.tight_layout()
```

```python
plt.figure(figsize=[4, 16])
plt.suptitle("My Subplots", size=20)

plt.subplot(4, 1, 1)
plt.plot(nums, color="#FF731D", label='x', marker='p')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(nums**2, color="#1746A2", label='x^2', marker='p')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(nums**3, color="#A66CFF", label='x^3', marker='p')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(nums**4, color="#31E1F7", label='x^4', marker='p')
plt.legend()

plt.tight_layout()
```
